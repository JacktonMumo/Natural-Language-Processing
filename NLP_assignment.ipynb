```python
import nltk
```

```python
from nltk import PunktSentenceTokenizer
import requests
from bs4 import BeautifulSoup
from nltk.probability import FreqDist
URL = "https://www.standardmedia.co.ke/health/health-science/article/2001496599/kenya-joins-international-vaccines-initiative-to-boost-healthcare"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
div = soup.findAll("div", class_ = "story-body")
soup = BeautifulSoup(str(div))
text = soup.get_text()
text


```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
token_words = word_tokenize(text)
token_words
```

```python
from nltk.corpus import stopwords
print(stopwords.fileids()) 
mystopwords = set(stopwords.words('english')) 
mystopwords
```

```python
filtered_words =[]
for word in token_words:
    if word.lower() not in mystopwords:
        filtered_words.append(word)
print("the new words without the common words is:",filtered_words)
```

```python
clean_words = [word for word in filtered_words if word.isalpha()]
clean_words
```

```python

frequency = FreqDist(clean_words)
frequency.most_common(30)
```
